### Dividing Data in Train test and valldation

```python
x_train=feature[:500]
y_train=label[:500]
x_val=feature[500:535]
y_val=label[500:535]
x_test=feature[535:]
y_test=label[535:]
```

## Creating Neural Network(Deep learning model-keras)

```python
from keras.models import Sequential
from keras.layers import Dense
model=Sequential()#For adding neural layers in sequential order
```

```python
#ADDING LAYERS OF NEURAL NETWORK
model.add(Dense(15,activation='relu',input_dim=30))#15 nodes in 1st layer and 30 input nodes(1st Hidden Layer)
model.add(Dense(8,activation='relu'))#2nd hidden layer
model.add(Dense(1,activation='sigmoid'))#output layer(sigmoid actiavtion bcoz we have output b/w 0 and 1)
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(x_train,y_train,epochs=10,batch_size=5,validation_data=(x_val,y_val))
```

```python
model.evaluate(x_test,y_test)
```

```python

model.evaluate(x_val,y_val)
```

### PREDICTION

```python
list_pred=[]
for i in range(30):
  l=[]
  sample = x_test[i]
  sample = np.reshape(sample, (1,30))

  if (model.predict(sample)[0][0] > 0.5):
    l.append("Benign")
  else:
    l.append("Malignant")

  if (y_test[i] == 1):
    l.append("Benign")
  else:
    l.append("Malignant")
  list_pred.append(l)
pred_data=pd.DataFrame(list_pred,columns=['Predicted Result','Actual Result'])
pred_data
```
